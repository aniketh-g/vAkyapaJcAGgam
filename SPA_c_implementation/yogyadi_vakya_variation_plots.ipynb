```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

sid_vakya_list = pd.read_csv('./tables/siddhanta_yogyadi_vakyas.csv') #vakya-s as predicted in siddhanta

vakya_list_files = ['./yogyadi_tables/spa_yogyadi_vakyas_%sAD.csv'%year for year in range(2023,2083)]
ref_vakya_list = sid_vakya_list

for vakya_list_file in vakya_list_files:
    vakya_list = pd.read_csv(vakya_list_file)
    plt.plot(range(48), np.subtract(vakya_list['true_long (na)'],ref_vakya_list['true_long (na)']), '.', label='%s'%vakya_list_file)

plt.title('Recent variation of yogyadivakyas wrt original vakyas')
plt.ylabel('difference between observed vakya and siddhanta vakya (degrees)')
plt.xlabel('vakya #')
plt.xticks(range(48))
fig = plt.gcf()
fig.set_size_inches(18.5, 8.5)
plt.legend()
plt.show()
```
