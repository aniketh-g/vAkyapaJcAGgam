```python
import pandas as pd
import astropy
from astropy import units as u
from astropy.coordinates import Angle
import matplotlib.pyplot as plt
import numpy as np

stel2023 = pandas.read_csv('./ephemeris_tables/stellarium_2023AD.csv')
nasa2023 = pandas.read_csv('./ephemeris_tables/horizons_2023AD.csv')

stelras = [Angle(a).degree for a in stel2023[' RA (J2000)']]
nasaras = [Angle(a*u.deg).degree for a in nasa2023['RA']]

print(stelras)
print(nasaras)


plt.plot(np.subtract(stelras,nasaras))
```
