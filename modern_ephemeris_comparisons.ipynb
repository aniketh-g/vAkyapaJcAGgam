```python
import pandas as pd
import astropy
from astropy import units as u
from astropy.coordinates import Angle
from astropy.coordinates import SkyCoord
from astropy.time import Time
import matplotlib.pyplot as plt
from matplotlib.pyplot import plot
import numpy as np

from astropy.coordinates import solar_system_ephemeris

stel2023 = pd.read_csv('./ephemeris_tables/stellarium_2023AD.csv')
nasa2023 = pd.read_csv('./ephemeris_tables/horizons_2023AD.csv')

stel_coords = [SkyCoord(ra=Angle(stel2023[' RA (J2000)'][i]), dec=Angle(stel2023[' Dec (J2000)'][i]), frame='icrs') for i in range(len(stel2023[' RA (J2000)']))]
jpl_coords = [SkyCoord(ra=nasa2023['RA'][i], dec=nasa2023['DEC'][i], frame='icrs', unit='deg') for i in range(len(nasa2023['RA']))]

stel_longs = [s.geocentrictrueecliptic.lon.degree for s in stel_coords]
jpl_longs = [j.geocentrictrueecliptic.lon.degree for j in jpl_coords]
plot(stel_longs)
plot(jpl_longs)
# print(jpl_coords)
# plt.plot(np.subtract(stelras,jpl_coords))
```
